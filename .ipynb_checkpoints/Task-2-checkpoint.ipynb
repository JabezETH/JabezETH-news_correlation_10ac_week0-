```python

def remove_stops(text, stops):
    words = text.split()
    final = []
    for word in words:
        if word not in stops:
            final.append(word)
    final = " ".join(final)
    final = final.translate(str.maketrans("", "", string.punctuation))
    final = "".join([i for i in final if not i.isdigit()])
    while " " in final:
        final = final.replace("  ", " ")
    return(final)
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download stopwords if not already downloaded
nltk.download('stopwords')
nltk.download('punkt')


# Function to preprocess text (tokenization and stopword removal)
def preprocess_text(text):
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return filtered_tokens

# Calculate similarity between two paragraphs
def calculate_similarity(paragraph1, paragraph2):
    tokens1 = preprocess_text(paragraph1)
    tokens2 = preprocess_text(paragraph2)
    similarity = nltk.jaccard_distance(set(tokens1), set(tokens2))
    return 1 - similarity

# Compare paragraphs and display results
for i in range(len(data)):
    for j in range(i + 1, len(data)):
        title1 = data.iloc[i]['title']
        content1 = data.iloc[i]['content']
        title2 = data.iloc[j]['title']
        content2 = data.iloc[j]['content']
        
        similarity = calculate_similarity(content1, content2)
        print("Similarity between '{}' and '{}' (Content): {:.2f}".format(title1, title2, similarity))
```

```python
def similarity(value, value2):
    tokens1 = word_tokenize(value)
    tokens2 = word_tokenize(value2)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens1 = [word for word in tokens1 if word.lower() not in stop_words]
    filtered_tokens2 = [word for word in tokens2 if word.lower() not in stop_words]

    # Calculate similarity using Jaccard similarity
    similarity = nltk.jaccard_distance(set(filtered_tokens1), set(filtered_tokens2))

    # Display results
    print("Title for Article 1", filtered_tokens1)
    print("Content For Article 1:", filtered_tokens2)
    
    print("Similarity:", 1 - similarity)
```

```python
from nltk.stem import WordNetLemmatizer


# Define a function to lemmatize text
def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    tokens = text.split()  # Assuming the title column contains text
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(lemmatized_tokens)

# Apply lemmatization to the 'title' column
lam_data = data['title'] = data['title'].apply(lemmatize_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialising the tfidf vectorizer with the default stopword list 
tfidf = TfidfVectorizer(stop_words="english", max_features= 1000, max_df = 0.5, smooth_idf=True)

#Vectorizing 'X' column
vector =tfidf.fit_transform(lam_data)

#Converting vector into an array
X= vector.toarray()
pd.DataFrame(X)
```

```python
from sklearn.decomposition import TruncatedSVD
svd_model = TruncatedSVD(n_components=5, algorithm='randomized', n_iter=100, random_state=122)
svd_model.fit(X)
print("The number of topics chosen are",len(svd_model.components_))
```

```python
svd_model.components_.shape
```

```python
terms = tfidf.get_feature_names()
topics = []

for i, comp in enumerate(svd_model.components_):
    terms_comp = zip(terms, comp)
    sorted_terms = sorted(terms_comp, key=lambda x:x[1], reverse=True)[:7]
    topics.append("Topic "+str(i)+": ")
    for term in sorted_terms:
        topics.append(term[0])

final_topic_list = [topics[i:i+8] for i in range(0, len(topics), 8)]

for x in final_topic_list:
    print (x)
```
